# Hybrid Intelligent System for Student Performance Prediction

## Objective
Develop a hybrid intelligent system combining:

- Fuzzy Logic
- Neural Network

To predict student performance level based on:

- Attendance
- Assignment Marks
- Test Marks

## Input Variables

### 1. Attendance
Range: 0 to 100

### 2. Assignment Marks
Range: 0 to 100

### 3. Test Marks
Range: 0 to 100

## Output Variable

### Performance Level

- Poor
- Average
- Good

## Concept of Hybrid Intelligent System

This system combines fuzzy logic and neural networks.

### Fuzzy Logic Role:
Handles uncertainty using linguistic values such as:
- Low
- Medium
- High

### Neural Network Role:
Learns patterns from training data and adjusts membership functions automatically.

This combination is called ANFIS (Adaptive Neuro Fuzzy Inference System).

## Sample Fuzzy Rules

1. If Attendance is Low and Test is Low then Performance is Poor  
2. If Attendance is Medium and Assignment is Medium then Performance is Average  
3. If Attendance is High and Test is High then Performance is Good  
4. If Attendance is High and Assignment is High then Performance is Good  
5. If Attendance is Low and Assignment is High then Performance is Average  
6. If Attendance is Medium and Test is High then Performance is Good

```python
clc;
clear;
close all;
```

```python
% Training Data
% [Attendance Assignment Test Output]

data = [
60 65 70 2;
85 80 90 3;
40 45 35 1;
75 70 80 3;
55 60 58 2;
30 40 38 1;
90 95 92 3;
65 68 66 2;
45 50 48 1;
78 82 79 3;
];
```

```python
% Generate Initial FIS using Grid Partitioning
fis = genfis1(data,3,'gbellmf');
```

```python
% Train ANFIS Model
epoch_n = 50;

[trnFis,error] = anfis(data,fis,epoch_n);
```

```python
% Plot Training Error
figure;
plot(error,'LineWidth',2);
xlabel('Epoch');
ylabel('Error');
title('ANFIS Training Error');
grid on;
```

```python
% View Membership Functions
figure;
plotmf(trnFis,'input',1);
title('Attendance Membership Functions');

figure;
plotmf(trnFis,'input',2);
title('Assignment Membership Functions');

figure;
plotmf(trnFis,'input',3);
title('Test Membership Functions');
```

```python
% Testing New Student Data
% [Attendance Assignment Test]

student = [82 78 88];

result = evalfis(student,trnFis);

disp(result);
```

## Output Interpretation

- Output near 1 = Poor
- Output near 2 = Average
- Output near 3 = Good

## Integration of Fuzzy Logic and Neural Network

### Step 1: Fuzzy Logic
Uses fuzzy IF-THEN rules to represent expert knowledge.

### Step 2: Neural Network Learning
ANFIS uses training data to tune:

- Membership function shapes
- Rule weights
- Output parameters

### Step 3: Final Prediction
After training, system predicts student performance more accurately.

## Advantages

- Handles uncertainty
- Learns automatically
- Better prediction accuracy
- Easy interpretation
