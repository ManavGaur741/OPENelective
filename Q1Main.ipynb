# Fuzzy Logic Washing Machine Cycle Time System

## Objective
Design a fuzzy logic system to determine the washing machine cycle time based on:

- Dirt Level
- Load Size

Using MATLAB Fuzzy Logic Toolbox.

## Input Variables

### 1. Dirt Level
Range: 0 to 10

Fuzzy Sets:
- Low
- Medium
- High

### 2. Load Size
Range: 0 to 10

Fuzzy Sets:
- Small
- Medium
- Large

## Output Variable

### Wash Cycle Time
Range: 0 to 60 minutes

Fuzzy Sets:
- Short
- Medium
- Long

## Fuzzy Rules

1. If Dirt is Low and Load is Small then Time is Short  
2. If Dirt is Low and Load is Medium then Time is Medium  
3. If Dirt is Low and Load is Large then Time is Medium  
4. If Dirt is Medium and Load is Small then Time is Medium  
5. If Dirt is Medium and Load is Large then Time is Long  
6. If Dirt is High and Load is Small then Time is Medium  
7. If Dirt is High and Load is Medium then Time is Long  
8. If Dirt is High and Load is Large then Time is Long

```python
fis = mamfis('Name','WashingMachine');
```

```python
% Input 1: Dirt Level
fis = addInput(fis,[0 10],'Name','Dirt');

fis = addMF(fis,'Dirt','trimf',[0 0 5],'Name','Low');
fis = addMF(fis,'Dirt','trimf',[0 5 10],'Name','Medium');
fis = addMF(fis,'Dirt','trimf',[5 10 10],'Name','High');
```

```python
% Input 2: Load Size
fis = addInput(fis,[0 10],'Name','Load');

fis = addMF(fis,'Load','trimf',[0 0 5],'Name','Small');
fis = addMF(fis,'Load','trimf',[0 5 10],'Name','Medium');
fis = addMF(fis,'Load','trimf',[5 10 10],'Name','Large');
```

```python
% Output: Wash Time
fis = addOutput(fis,[0 60],'Name','Time');

fis = addMF(fis,'Time','trimf',[0 0 30],'Name','Short');
fis = addMF(fis,'Time','trimf',[15 30 45],'Name','Medium');
fis = addMF(fis,'Time','trimf',[30 60 60],'Name','Long');
```

```python
ruleList = [
1 1 1 1 1;
1 2 2 1 1;
1 3 2 1 1;
2 1 2 1 1;
2 3 3 1 1;
3 1 2 1 1;
3 2 3 1 1;
3 3 3 1 1;
];

fis = addRule(fis,ruleList);
```

```python
% Show Membership Functions
figure;
plotmf(fis,'input',1);
title('Dirt Level Membership Functions');

figure;
plotmf(fis,'input',2);
title('Load Size Membership Functions');

figure;
plotmf(fis,'output',1);
title('Wash Time Membership Functions');
```

```python
% Open Rule Viewer
ruleview(fis)
```

```python
% Example Test
input = [7 8];
output = evalfis(fis,input);

disp(['Wash Cycle Time = ',num2str(output),' minutes']);
```

## Output

The fuzzy logic system calculates washing cycle time according to dirt level and load size using Mamdani inference system.
